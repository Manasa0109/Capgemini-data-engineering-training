```sql
drop table employee;
CREATE TABLE Employee (
emp_id INT PRIMARY KEY,
emp_name VARCHAR(50),
department VARCHAR(50),
salary DECIMAL(10, 2),
joining_date DATE
);

```

RANK() Based Questions

```sql
select emp_name,salary,rank() over(order by salary desc) as rn
from employee
```

```sql
select emp_name,department,salary,
   rank() over(partition by department order by  salary desc ) as rn
from employee
```

```sql
select emp_name,rank() over(order by joining_date desc) as rn
from employee
```

```sql
--14.	Rank employees within department based on salary (lowest first). 

select rank() over (partition by department order by salary) as rn,*
from employee
```

```sql
--15.	Rank employees based on name alphabetically. 
select rank() over(order by emp_name) as name_order,emp_name
from employee
```

DENSE_RANK() Based Questions

```sql
select emp_name,salary,dense_rank() over(order by salary desc)as rn
from employee
```

```sql
select emp_name,salary,dense_rank() over (partition by department order by salary desc) as rn
from employee
```

```sql
select emp_name,department,dense_rank() over(order by joining_date)
from employee
```

```sql
select emp_name,department,joining_date,dense_rank() over (partition by department order by joining_date) as rn
from employee
```
