```sql
select emp_name,salary,
CASE 
    WHEN experience >=8 and performance_rating='A' then salary*1.20
    when experience>=5 and performance_rating='B' then salary*1.15
    when performance_rating='C' then salary
    else salary*1.10
end as new_salary
from emp
```

```sql
select emp_name,department,performance_rating,
CASE 
  WHEN department='Finance' THEN 
  CASE
      WHEN performance_rating='A' THEN salary*0.2
      WHEN performance_rating='B' THEN salary*0.15
      WHEN performance_rating='C' THEN salary*0.5
  end
  when department='Engineering' THEN
  CASE 
       WHEN performance_rating='A' THEN salary*0.18
      WHEN performance_rating='B' THEN salary*0.12
      WHEN performance_rating='C' THEN salary*0.3
  end
  else salary*0.1
  end as bonus
  from emp

      
```

```sql
select emp_name,salary,performance_rating,
CASE 
    WHEN salary>80000 and performance_rating='A' then 'HIGH'
    WHEN salary between 50000 and 80000 or performance_rating='B' THEN 'MID'
    WHEN salary<50000 or performance_rating='C' THEN  'LOW'
END as Performer
from emp
```

```sql
select emp_name,department,experience,
CASE 
    when department='HR' then
        case
            when experience<5 then 'high'
            when experience>5 then 'low'
            else 'medium'
        end 
    when department='Engineering' or department='Finance' then
        case
            when experience>8 then 'low'
            when experience<=8 then 'medium'
        end
    else 'Medium'
end as Risk_Level
from emp

```

```sql
select emp_name,
CASE 
    WHEN performance_rating='A' THEN 
    CASE 
        WHEN salary>80000 THEN 
        CASE 
            WHEN experience>5 THEN salary*0.25 
            else salary*0.2
        end
        WHEN salary between 50000 and 80000 THEN salary*0.15
    end 
    WHEN performance_rating='B' THEN 
    CASE 
        WHEN experience>5 THEN salary*0.12 
        else salary*0.1 
    end 
    when performance_rating='C' THEN salary
end as hike 
from emp
```
