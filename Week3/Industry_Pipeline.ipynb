```python
clean_df = customers.join(policies, on="customer_id", how="inner").join(claims, on="policy_id",how="left").join(policy_agent, on="policy_id",   how="left").join(agents,  on="agent_id", how="left")

clean_df = clean_df.select(
    col("policy_id"),
    col("customer_id"),
    customers["name"].alias("customer_name"), 
    col("age"),
    col("city"),
    col("agent_id"),
    agents["name"].alias("agent_name"),     
    col("region"),
    col("policy_type"),
    col("premium"),
    col("start_date"),
    col("claim_id"),
    col("claim_amount"),
    col("claim_date"),
    col("status"),
    policies["Data Validation"].alias("Premium_Validation"),
    claims["Data Validation"].alias("Claim_Validation")
)

clean_df.orderBy("customer_id").display()
```

```python
clean_df = clean_df.fillna({
    "agent_id":     0,
    "agent_name":   "Unknown",
    "region":       "Unknown",
    "claim_id":     0,
    "claim_amount": 0,
    "status":       "Unknown",
    "Claim_Validation":"Unknown"
})
clean_df = clean_df.withColumn(
    "claim_date",
    when(col("claim_date").isNull(), to_date(lit("2023-01-01")))
    .otherwise(col("claim_date"))
)

clean_df.orderBy("customer_id").display()
```

```python
premium_per_customer=clean_df.groupBy("customer_id").agg(
    sum(
        when(col("Premium_Validation")!="Invalid",col("premium")).otherwise(0)
).alias("Premium per Customer"))
premium_per_customer.orderBy("customer_id").display()
```

```python
claim_per_customer=clean_df.groupBy("customer_id").agg(
    sum(
        when(col("Claim_Validation")!="Invalid",col("claim_amount")).otherwise(0)
    ).alias("Claim per Customer")
)
claim_per_customer.orderBy("customer_id").display()
```

```python
compute_score=claim_per_customer.join(premium_per_customer,on="customer_id",how="inner")
compute_score=compute_score.withColumn("Score",
                                    when(col("Premium per Customer")>0,round(col("Claim per Customer")/col("Premium per Customer"),2)
                                    ).otherwise(0))
compute_score.display()
```

```python
clean_df.groupBy("city").agg(
    sum(
        when(col("Premium_Validation")!="Invalid",col("premium")).otherwise(0)
    ).alias("City wise premium"),
    sum(
        when(col("Claim_Validation")!="Invalid",col("claim_amount")).otherwise(0)
    ).alias("City wise Claim")
).display()
```

```python
agents_prem=clean_df.groupBy("agent_id").agg(
    sum(
        when(col("Premium_Validation")!="Invalid",col("premium")).otherwise(0)
    ).alias("Agent wise Premium")
)
agents_prem.display()
```

```python
window=Window.orderBy(col("Agent wise Premium").desc())
rank_agents=agents_prem.select("agent_id","Agent wise Premium",rank().over(window).alias("Ranking Agents"))
rank_agents.display()
```

```python
spark.sql(
    """
    with customer_premium as (
        select customer_id,sum(premium) as total_premium
         from policy
         where premium>0
         group by customer_id
         ),
         customer_claims as (
             select p.customer_id,sum(c.claim_amount)as total_claim
             from claims c join policy p
             on c.policy_id=p.policy_id
             where c.claim_amount>0
             group by p.customer_id
             
        ),
        compute_risk_score as(
            select cp.customer_id,
            cp.total_premium,
            cc.total_claim,
            round(cc.total_claim/cp.total_premium,2) as risk
            from customer_premium cp join customer_claims cc
            on cp.customer_id=cc.customer_id
            where cp.total_premium>0 and total_claim>0
        )
        select crs.customer_id,crs.risk,dense_rank() over(order by risk desc) as rank
        from compute_risk_score crs
        limit 3;
    
    """
).display()
```

```python
spark.sql(
    """
    with city_claim as(
        select p.customer_id,
        month(c.claim_date) as claim_month,
        sum(c.claim_amount) as total_claim
       
        from policy p join claims c 
        on p.policy_id=c.policy_id
        where c.claim_amount>0
        group by p.customer_id,month(c.claim_date)
    ),
    lagged_claim as(
        select customer_id,claim_month,total_claim,
        lag(total_claim,1) over(partition by(customer_id) order by (claim_month)) as previous_claim
        from city_claim

    )
    select customer_id,total_claim,previous_claim
    from lagged_claim 
    where total_claim>previous_claim
    """
).display()
```
